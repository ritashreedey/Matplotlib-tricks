# matplotlib exercise

```python
import numpy as np
```

```python
x=np.arange(0,100)
y=x*2
z=x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#create a figure object called fig using plt.figure()
#add axes
#plot x,y and add labels
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
#add two axes as ax1 and ax2
```

```python
fig=plt.figure()
axes1=fig.add_axes([0,0,1,1])
axes2=fig.add_axes([0.2,0.5,0.2,0.2])
```

```python
#now plot x,y on both axes
```

```python
fig=plt.figure()
axes1=fig.add_axes([0,0,1,1])
axes2=fig.add_axes([0.2,0.5,0.2,0.2])
axes1.plot(x,y,color='red')
axes2.plot(x,y,color='green')
```

```python
#create a plot using axes at[0,0,1,1] and [0.2,0.5,0.4,0.4]
```

```python
fig=plt.figure()
axes1=fig.add_axes([0,0,1,1])
axes2=fig.add_axes([0.2,0.5,0.4,0.4])
```

```python
#now use x,y,z arrays to recreate the plot
```

```python
fig=plt.figure()
axes1=fig.add_axes([0,0,1,1])
axes2=fig.add_axes([0.2,0.5,0.4,0.4])

axes1.plot(x,z)
axes1.set_xlabel('X')
axes1.set_ylabel('Z')

axes2.plot(x,y)
axes2.set_title('Zoom')
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_xlim([20,22])
axes2.set_ylim([30,50])
```

```python
#use plt.subplots(nrows=1,ncols=2)
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y,color='red',lw=3,ls='--')
axes[1].plot(x,z,color='purple',lw=3,ls='-')

plt.tight_layout()
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(12,2))

axes[0].plot(x,y,color='red',lw=3,ls='--')
axes[1].plot(x,z,color='purple',lw=3,ls='-')

plt.tight_layout()
```
